```python
df_inv['min_available_qty'] = (df_inv['sell_through'] * 30).round(2)
df_inv['max_available_qty'] = (df_inv['sell_through'] * 60).round(2)

import numpy as np

conditions = [
    (df_inv['Available Inventory'] > df_inv['min_available_qty']) &
    (df_inv['Available Inventory'] <= df_inv['max_available_qty']),
    
    (df_inv['Available Inventory'] > df_inv['max_available_qty']),
    
    (df_inv['Available Inventory'] <= df_inv['min_available_qty'])
]

choices = [
    'Healthy',
    'Overstock',
    'Understock'
]

df_inv['Stock_Status'] = np.select(conditions, choices, default='Unknown')
```

```python
import pandas as pd

# -------------------------------
# STEP 1: Prepare df_ledger
# -------------------------------

# Temporary datetime column (keeps original Date intact)
df_ledger["Date_temp"] = pd.to_datetime(df_ledger["Date"], format="%d-%b-%Y")

# Create year_month column
df_ledger["year_month"] = df_ledger["Date_temp"].dt.to_period("M")

# Ensure ASIN column name matches
df_ledger.rename(columns={'ASIN': 'asin'}, inplace=True)

# -------------------------------
# STEP 2: Merge min_available_qty
# -------------------------------

df_ledger = df_ledger.merge(
    df_inv[['asin', 'min_available_qty']],
    on='asin',
    how='left'
)

# -------------------------------
# STEP 3: Create in-stock flag
# -------------------------------

df_ledger['in_stock_flag'] = (
    df_ledger['Ending Warehouse Balance'] > df_ledger['min_available_qty']
).astype(int)

# -------------------------------
# STEP 4: Calculate Instock Days
# -------------------------------

# Days in stock
days_instock = (
    df_ledger[df_ledger['in_stock_flag'] == 1]
    .groupby(['asin', 'year_month'])['Date_temp']
    .nunique()
    .reset_index(name='days_instock')
)

# Total days
total_days = (
    df_ledger.groupby(['asin', 'year_month'])['Date_temp']
    .nunique()
    .reset_index(name='total_days')
)

# Merge both
Instock_Rate_df = days_instock.merge(
    total_days,
    on=['asin', 'year_month'],
    how='right'
)

# -------------------------------
# STEP 5: Calculate Instock Rate
# -------------------------------

Instock_Rate_df['Instock Rate'] = (
    Instock_Rate_df['days_instock'].fillna(0) /
    Instock_Rate_df['total_days'] * 100
).round(2)

# Keep only required columns
Instock_Rate_df = Instock_Rate_df[['asin', 'year_month', 'Instock Rate']]

# -------------------------------
# STEP 6: Prepare df_inv for merge
# -------------------------------
df_inv["Date_temp"] = pd.to_datetime(
    df_inv["Date"],
    format='mixed',
    dayfirst=True,
    errors='coerce'   # prevents crash
)
df_inv["year_month"] = df_inv["Date_temp"].dt.to_period("M")

# -------------------------------
# STEP 7: Correct Merge (FIXED)
# -------------------------------

df_inv = df_inv.merge(
    Instock_Rate_df,
    on=['asin', 'year_month'],   # ✅ FIX: include year_month
    how='left'
)

# -------------------------------
# STEP 8: Fill missing values
# -------------------------------

df_inv['Instock Rate'] = df_inv['Instock Rate'].fillna(0)

# -------------------------------
# STEP 9: Cleanup
# -------------------------------

df_ledger.drop(columns=["Date_temp", "year_month"], inplace=True)
df_inv.drop(columns=["Date_temp", "year_month"], inplace=True)
```

```python
import os
import pandas as pd
from datetime import datetime, timedelta

base_path = r"C:\Users\Shubham Verma\OneDrive - Incircle Ecoproducts Pvt Ltd\Thrivebrands - Client Database\Pattex\seller_central\uae\input\revenue\orders"

today = datetime.today()
day = today.day
year = today.year
month = today.month

# -------- MONTH MAP --------
month_map = {
    1:"jan",2:"feb",3:"mar",4:"apr",5:"may",6:"jun",
    7:"jul",8:"aug",9:"sep",10:"oct",11:"nov",12:"dec"
}

# -------- PREVIOUS MONTH --------
prev_date = today.replace(day=1) - timedelta(days=1)
prev_month = prev_date.month
prev_year = prev_date.year

# -------- DECIDE FOLDERS --------
folders_to_read = []

# Always include current month
folders_to_read.append((year, month))

# Include previous month if day ≤ 3
if day <= 3:
    folders_to_read.append((prev_year, prev_month))

print("Folders to read:", folders_to_read)

df_list = []

# -------- LOOP THROUGH SELECTED FOLDERS --------
for yr, mon in folders_to_read:

    folder_path = os.path.join(base_path, str(yr), month_map[mon])

    if not os.path.exists(folder_path):
        continue

    for file in os.listdir(folder_path):

        if file.endswith(".txt"):

            file_path = os.path.join(folder_path, file)

            df = pd.read_csv(file_path, sep="\t")
            df_list.append(df)

# -------- COMBINE --------
df_orders = pd.concat(df_list, ignore_index=True) if df_list else pd.DataFrame()

# -------- FILTER --------
if not df_orders.empty:
    df_orders = df_orders[
        df_orders['product-name'].str.contains('pattex|pritt', case=False, na=False)
    ]

    # -------- DATE FORMAT --------
    df_orders["purchase-date"] = pd.to_datetime(
        df_orders["purchase-date"], errors="coerce"
    ).dt.strftime("%d-%b-%Y")

print("Total orders rows:", len(df_orders))
```

```python
import os
import pandas as pd
from datetime import datetime, timedelta

base_path = r"C:\Users\Shubham Verma\OneDrive - Incircle Ecoproducts Pvt Ltd\Thrivebrands - Client Database\Pattex\seller_central\uae\input\ads\SPAP"

today = datetime.today()
day = today.day
year = today.year
month = today.month

# -------- MONTH MAP --------
month_map = {
    1:"jan",2:"feb",3:"mar",4:"apr",5:"may",6:"jun",
    7:"jul",8:"aug",9:"sep",10:"oct",11:"nov",12:"dec"
}

# -------- PREVIOUS MONTH --------
prev_date = today.replace(day=1) - timedelta(days=1)
prev_month = prev_date.month
prev_year = prev_date.year

# -------- DECIDE FOLDERS --------
folders_to_read = []

# Always include current month
folders_to_read.append((year, month))

# Include previous month if day ≤ 3
if day <= 3:
    folders_to_read.append((prev_year, prev_month))

print("Folders to read:", folders_to_read)

df_list = []

# -------- LOOP --------
for yr, mon in folders_to_read:

    folder_path = os.path.join(base_path, str(yr), month_map[mon])

    if not os.path.exists(folder_path):
        continue

    for file in os.listdir(folder_path):

        if file.endswith(".xlsx"):

            file_path = os.path.join(folder_path, file)

            df = pd.read_excel(file_path)
            df_list.append(df)

# -------- COMBINE --------
df_spap = pd.concat(df_list, ignore_index=True) if df_list else pd.DataFrame()

# -------- FILTER --------
if not df_spap.empty:
    df_spap = df_spap[
        df_spap['Portfolio name'].str.contains('pattex|pritt', case=False, na=False)
    ]

    # -------- DATE FORMAT --------
    df_spap["Date"] = pd.to_datetime(
        df_spap["Date"],
        format="%m/%d/%Y",
        errors="coerce"
    ).dt.strftime("%d-%b-%Y")

print("Total SPAP rows:", len(df_spap))
```
