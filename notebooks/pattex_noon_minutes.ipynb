## Read Inventory 

```python
import pandas as pd
from pathlib import Path

# Root folder
root_folder = Path(
    r"C:\Users\Shubham Verma\OneDrive - Incircle Ecoproducts Pvt Ltd\Thrivebrands - Client Database\Pattex\noon_minutes\uae\input\inventory"
)

# Find all csv and excel files recursively
files = list(root_folder.rglob("*.csv")) + \
        list(root_folder.rglob("*.xlsx")) + \
        list(root_folder.rglob("*.xls"))

print(f"Total files found: {len(files)}")

dfs = []

for file in files:
    try:
        if file.suffix.lower() == ".csv":
            temp_df = pd.read_csv(file)
        else:
            temp_df = pd.read_excel(file)

        # Optional metadata
        temp_df["source_file"] = file.name
        temp_df["month"] = file.parent.name          # june, may, etc.
        temp_df["year"] = file.parent.parent.name    # 2026, 2027, etc.
        temp_df["file_path"] = str(file)

        dfs.append(temp_df)

        print(f"✓ Loaded: {file.name}")

    except Exception as e:
        print(f"✗ Error reading {file}: {e}")

# Combine all files
df = pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()

print("\nFinal Shape:", df.shape)

# Preview
df
```

## Select Cols

```python
df.rename(columns={"Refresh Time": "Date", "Stock": "Inventory", }, inplace=True)
cols_to_keep = [
    "Date",
    "Inventory",
    "ZSKU",
]
df = df[cols_to_keep]
df
```

## GroupBy the data

```python
import pandas as pd

# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'])

# Format date as 01-June-2026
df['Date'] = df['Date'].dt.strftime('%d-%B-%Y')

# Group by Date and ZSKU and sum Inventory
df_inv = df.groupby(['Date', 'ZSKU'], as_index=False)['Inventory'].sum()

df_inv
```

## Add product details

## last 30 day sale

```python
import pandas as pd

# Convert dates to datetime
df_inv['Date'] = pd.to_datetime(df_inv['Date'])
df_sales['Date'] = pd.to_datetime(df_sales['Date'])

def get_last_30_day_sales(row):
    # Exclude current day and previous day
    start_date = row['Date'] - pd.Timedelta(days=31)
    end_date = row['Date'] - pd.Timedelta(days=2)

    return df_sales.loc[
        (df_sales['ZSKU'] == row['ZSKU']) &
        (df_sales['Date'] >= start_date) &
        (df_sales['Date'] <= end_date),
        'qty_sold'
    ].sum()

# Create new column
df_inv['last_30_day_sale_qty'] = df_inv.apply(
    get_last_30_day_sales,
    axis=1
)
```

## DOS

```python
import numpy as np

df_inv['DOS'] = np.where(
    df_inv['last_30_day_sale_qty'] > 0,
    df_inv['Inventory'] * 30 / df_inv['last_30_day_sale_qty'],
    np.nan
)
```

## Marketing

```python
import pandas as pd
from pathlib import Path

# Root folder
root_folder = Path(
    r"C:\Users\Shubham Verma\OneDrive - Incircle Ecoproducts Pvt Ltd\Thrivebrands - Client Database\Pattex\noon_minutes\uae\input\marketing"
)

# Find all csv and excel files recursively
files = list(root_folder.rglob("*.csv")) + \
        list(root_folder.rglob("*.xlsx")) + \
        list(root_folder.rglob("*.xls"))

print(f"Total files found: {len(files)}")

dfs = []

for file in files:
    try:
        if file.suffix.lower() == ".csv":
            temp_df = pd.read_csv(file)
        else:
            temp_df = pd.read_excel(file)

        # Optional metadata
        temp_df["source_file"] = file.name
        temp_df["month"] = file.parent.name          # june, may, etc.
        temp_df["year"] = file.parent.parent.name    # 2026, 2027, etc.
        temp_df["file_path"] = str(file)

        dfs.append(temp_df)

        print(f"✓ Loaded: {file.name}")

    except Exception as e:
        print(f"✗ Error reading {file}: {e}")

# Combine all files
df_marketing = pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()

print("\nFinal Shape:", df_marketing.shape)

# Preview
df_marketing
```

```python
cols_to_keep = [
    "Date",
    "SKU",
    "Clicks",
    "Spends",
    "Revenue",
    "Orders"
]
df_marketing = df_marketing[cols_to_keep]
df_marketing
```

```python
import pandas as pd

# Convert Date column to datetime
df_marketing['Date'] = pd.to_datetime(df_marketing['Date'])

# Columns to be summed
sum_cols = ['Clicks', 'Spends', 'Revenue', 'Orders']

# Remove commas, convert to numeric, and handle invalid values
for col in sum_cols:
    df_marketing[col] = (
        df_marketing[col]
        .astype(str)
        .str.replace(',', '', regex=False)
    )
    df_marketing[col] = pd.to_numeric(df_marketing[col], errors='coerce').fillna(0)

# Group by Date and SKU and sum the specified columns
df_marketing = (
    df_marketing
    .groupby(['Date', 'SKU'], as_index=False)[sum_cols]
    .sum()
)

# Format date as 01-June-2026
df_marketing['Date'] = df_marketing['Date'].dt.strftime('%d-%B-%Y')

df_marketing.to_csv(r"C:\Users\Shubham Verma\OneDrive - Incircle Ecoproducts Pvt Ltd\Thrivebrands - Client Database\Pattex\noon_minutes\uae\output\marketing\noon_minutes_marketing.csv", index=False)
```
