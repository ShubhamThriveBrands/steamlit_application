## Read Inventory 

```python
import pandas as pd
from pathlib import Path

# Root folder
root_folder = Path(
    r"C:\Users\Shubham Verma\OneDrive - Incircle Ecoproducts Pvt Ltd\Thrivebrands - Client Database\Pattex\noon_minutes\uae\input\inventory"
)

# Find all csv and excel files recursively
files = list(root_folder.rglob("*.csv")) + \
        list(root_folder.rglob("*.xlsx")) + \
        list(root_folder.rglob("*.xls"))

print(f"Total files found: {len(files)}")

dfs = []

for file in files:
    try:
        if file.suffix.lower() == ".csv":
            temp_df = pd.read_csv(file)
        else:
            temp_df = pd.read_excel(file)

        # Optional metadata
        temp_df["source_file"] = file.name
        temp_df["month"] = file.parent.name          # june, may, etc.
        temp_df["year"] = file.parent.parent.name    # 2026, 2027, etc.
        temp_df["file_path"] = str(file)

        dfs.append(temp_df)

        print(f"✓ Loaded: {file.name}")

    except Exception as e:
        print(f"✗ Error reading {file}: {e}")

# Combine all files
df = pd.concat(dfs, ignore_index=True) if dfs else pd.DataFrame()

print("\nFinal Shape:", df.shape)

# Preview
df
```

```python
df.rename(columns={"Refresh Time": "Date", "Stock": "Inventory", }, inplace=True)
cols_to_keep = [
    "Date",
    "Inventory",
    "ZSKU",
    "Warehouse Code",
]
df = df[cols_to_keep]
df
```

## last 30 day sale

```python
import pandas as pd

# Ensure dates are datetime
df_inv['Date'] = pd.to_datetime(df_inv['Date'])
df_sales['Date'] = pd.to_datetime(df_sales['Date'])

# Create a copy with only required columns
sales = df_sales[['Date', 'ZSKU', 'Warehouse', 'qty_sold']].copy()

# Function to calculate last 30 day sales
def get_last_30_day_sales(row):
    end_date = row['Date'] - pd.Timedelta(days=2)
    start_date = row['Date'] - pd.Timedelta(days=31)

    mask = (
        (sales['ZSKU'] == row['ZSKU']) &
        (sales['Warehouse'] == row['Warehouse']) &
        (sales['Date'] >= start_date) &
        (sales['Date'] <= end_date)
    )

    return sales.loc[mask, 'qty_sold'].sum()

# Add column
df_inv['last_30_day_sale'] = df_inv.apply(get_last_30_day_sales, axis=1)
```

## DOS

```python
import numpy as np

df_inv['DOS'] = np.where(
    df_inv['last_30_day_sale'] > 0,
    df_inv['Inventory'] * 30 / df_inv['last_30_day_sale'],
    np.nan
)
```

```python
df_inv['DOS'] = df_inv['DOS'].astype(float).round(2)
df_inv["Alert"] = np.where(
    df_inv['DOS'] < 10,
    "Low Stock",
    ""
)
```
