## Read Files

```python
import pandas as pd
import os
import re
import warnings

warnings.filterwarnings(
    "ignore",
    message="Workbook contains no default style"
)

# =====================================================
# CONFIG
# =====================================================

root_folder = r"C:\Users\Shubham Verma\OneDrive - Incircle Ecoproducts Pvt Ltd\Thrivebrands - Client Database\Pattex\Ads\Kinetica\UK\input"

date_pattern = re.compile(r'(\d{2}-\d{2}-\d{4})')

spap_dfs = []
sb_dfs = []
sd_dfs = []

files_processed = 0

# =====================================================
# LOOP THROUGH FILES
# =====================================================

for root, dirs, files in os.walk(root_folder):

    for file in files:

        if not file.lower().endswith(".xlsx"):
            continue

        file_path = os.path.join(root, file)

        print("\n" + "="*80)
        print(f"FILE : {file}")
        print("="*80)

        # -------------------------------------------------
        # DATE
        # -------------------------------------------------

        date_match = date_pattern.search(file)

        if not date_match:
            print("❌ Date not found in filename")
            continue

        file_date = pd.to_datetime(
            date_match.group(1),
            format="%d-%m-%Y"
        )

        try:

            files_processed += 1

            xls = pd.ExcelFile(file_path)

            print("Sheets Found:")
            print(xls.sheet_names)

            # =====================================================
            # SPONSORED PRODUCTS
            # =====================================================

            sp_sheet = next(
                (
                    s for s in xls.sheet_names
                    if "sponsored products" in s.lower()
                ),
                None
            )

            if sp_sheet:

                print(f"\nSP Sheet Found -> {sp_sheet}")

                df = pd.read_excel(
                    file_path,
                    sheet_name=sp_sheet
                )

                print("Rows before filter:", len(df))

                if "Entity" in df.columns:

                    print(
                        "Unique Entities:",
                        df["Entity"]
                        .astype(str)
                        .str.strip()
                        .unique()
                    )

                    mask = (
                        df["Entity"]
                        .astype(str)
                        .str.replace(r"\s+", " ", regex=True)
                        .str.strip()
                        .str.lower()
                        .eq("product ad")
                    )

                    df = df[mask].copy()

                    print("Rows after filter:", len(df))

                    if not df.empty:
                        df["Date"] = file_date
                        spap_dfs.append(df)

            else:
                print("❌ Sponsored Products sheet not found")

            # =====================================================
            # SPONSORED BRANDS
            # =====================================================

            sb_sheet = next(
                (
                    s for s in xls.sheet_names
                    if "sponsored brands" in s.lower()
                ),
                None
            )

            if sb_sheet:

                print(f"\nSB Sheet Found -> {sb_sheet}")

                df = pd.read_excel(
                    file_path,
                    sheet_name=sb_sheet
                )

                print("Rows before filter:", len(df))

                if "Entity" in df.columns:

                    print(
                        "Unique Entities:",
                        df["Entity"]
                        .astype(str)
                        .str.strip()
                        .unique()
                    )

                    mask = (
                        df["Entity"]
                        .astype(str)
                        .str.replace(r"\s+", " ", regex=True)
                        .str.strip()
                        .str.lower()
                        .eq("campaign")
                    )

                    df = df[mask].copy()

                    print("Rows after filter:", len(df))

                    if not df.empty:
                        df["Date"] = file_date
                        sb_dfs.append(df)

                else:
                    print("❌ Entity column not found")

            else:
                print("❌ Sponsored Brands sheet not found")

            # =====================================================
            # SPONSORED DISPLAY
            # =====================================================

            sd_sheet = next(
                (
                    s for s in xls.sheet_names
                    if "sponsored display" in s.lower()
                ),
                None
            )

            if sd_sheet:

                print(f"\nSD Sheet Found -> {sd_sheet}")

                df = pd.read_excel(
                    file_path,
                    sheet_name=sd_sheet
                )

                print("Rows before filter:", len(df))

                if "Entity" in df.columns:

                    print(
                        "Unique Entities:",
                        df["Entity"]
                        .astype(str)
                        .str.strip()
                        .unique()
                    )

                    mask = (
                        df["Entity"]
                        .astype(str)
                        .str.replace(r"\s+", " ", regex=True)
                        .str.strip()
                        .str.lower()
                        .eq("product ad")
                    )

                    df = df[mask].copy()

                    print("Rows after filter:", len(df))

                    if not df.empty:
                        df["Date"] = file_date
                        sd_dfs.append(df)

                else:
                    print("❌ Entity column not found")

            else:
                print("❌ Sponsored Display sheet not found")

        except Exception as e:

            print(f"\n❌ ERROR IN FILE")
            print(file_path)
            print(e)

# =====================================================
# CONCAT
# =====================================================

df_spap = (
    pd.concat(spap_dfs, ignore_index=True)
    if len(spap_dfs) > 0
    else pd.DataFrame()
)

df_sb = (
    pd.concat(sb_dfs, ignore_index=True)
    if len(sb_dfs) > 0
    else pd.DataFrame()
)

df_sd = (
    pd.concat(sd_dfs, ignore_index=True)
    if len(sd_dfs) > 0
    else pd.DataFrame()
)

# =====================================================
# SUMMARY
# =====================================================

print("\n" + "="*80)
print("SUMMARY")
print("="*80)

print("Files Processed :", files_processed)

print("SP Files Added :", len(spap_dfs))
print("SB Files Added :", len(sb_dfs))
print("SD Files Added :", len(sd_dfs))

print()

print("df_spap shape :", df_spap.shape)
print("df_sb shape   :", df_sb.shape)
print("df_sd shape   :", df_sd.shape)
```

## Handle Multiple Asins

```python
numeric_cols = [
    'Impressions',
    'Clicks',
    'Conversion Rate',
    'Spend',
    'Sales',
    'Orders',
    'Units',
    'CPC',
    'ROAS',
    'ACOS'
]

# Create a copy
df_sb_expanded = df_sb.copy()

# Convert ASIN string to list
df_sb_expanded['ASIN'] = (
    df_sb_expanded['ASIN']
    .fillna('')
    .astype(str)
    .str.split(r'\s*,\s*')
)

# Count ASINs per row
df_sb_expanded['asin_count'] = df_sb_expanded['ASIN'].apply(
    lambda x: len([i for i in x if i.strip()])
)

# Divide metrics by number of ASINs
for col in numeric_cols:
    if col in df_sb_expanded.columns:
        df_sb_expanded[col] = (
            df_sb_expanded[col] /
            df_sb_expanded['asin_count']
        )

# Create one row per ASIN
df_sb_expanded = df_sb_expanded.explode('ASIN')

# Remove helper column
df_sb_expanded = df_sb_expanded.drop(columns='asin_count')

# Clean ASIN values
df_sb_expanded['ASIN'] = df_sb_expanded['ASIN'].str.strip()

print(df_sb_expanded.shape)
df_sb_expanded
```
