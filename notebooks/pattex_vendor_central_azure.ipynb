```python
import os
import re
import pandas as pd
from datetime import datetime, timedelta
import calendar

# -------- BASE PATH --------
BASE_PATH = r"C:\Users\Shubham Verma\OneDrive - Incircle Ecoproducts Pvt Ltd\Thrivebrands - Client Database\Pattex\vendor_central\uae\sales\input"

dfs = []

# -------- REGEX FOR DATE EXTRACTION --------
pattern = re.compile(r"_(\d{2})_([A-Za-z]{3})_(\d{4})")

# -------- GET TODAY DATE --------
today = datetime.today()
day = today.day
current_month = today.month
current_year = today.year

# -------- FIND PREVIOUS MONTH --------
prev_date = today.replace(day=1) - timedelta(days=1)
prev_month = prev_date.month
prev_year = prev_date.year

# -------- MONTHS TO READ --------
months_to_read = []

if day <= 3:
    months_to_read.append((prev_year, prev_month))
    months_to_read.append((current_year, current_month))
else:
    months_to_read.append((current_year, current_month))

print("Months to read:", months_to_read)

# -------- LOOP THROUGH REQUIRED MONTHS --------
for year, month in months_to_read:

    year_path = os.path.join(BASE_PATH, str(year))

    if not os.path.exists(year_path):
        continue

    # convert month number → jan/feb/mar
    month_name = calendar.month_abbr[month].lower()

    month_path = os.path.join(year_path, month_name)

    if not os.path.exists(month_path):
        continue

    # -------- LOOP THROUGH FILES --------
    for file in os.listdir(month_path):

        if file.lower().endswith(".csv"):

            file_path = os.path.join(month_path, file)
            print("Reading:", file_path)

            # -------- APPLY SKIPROWS LOGIC --------
            if month_name in ["jan", "feb"]:
                df = pd.read_csv(file_path, skiprows=1)
            else:
                df = pd.read_csv(file_path)

            # -------- EXTRACT DATE FROM FILENAME --------
            match = pattern.search(file)

            if match:

                day_, month_str, year_str = match.groups()

                month_number = datetime.strptime(
                    month_str.title(), "%b"
                ).month

                date_value = datetime(
                    int(year_str),
                    int(month_number),
                    int(day_)
                ).date()

                df["Date"] = date_value

            else:
                print(f"⚠ Date not found in filename: {file}")
                df["Date"] = None

            dfs.append(df)

# -------- MERGE ALL FILES --------
if dfs:
    final_df = pd.concat(dfs, ignore_index=True)

    print("✅ All files merged successfully")
    print("Total files read:", len(dfs))
    print("Final dataframe shape:", final_df.shape)

else:
    print("❌ No CSV files found.")
```

### OOS Date

```python
import pandas as pd
from datetime import date

# Create temporary datetime column for calculations
inv_df["Date_temp"] = pd.to_datetime(inv_df["Date"], format="%d-%b-%Y")

today = date.today()
curr_month = today.month
curr_year = today.year


def calculate_oos(group):
    group = group.sort_values("Date_temp").copy()

    # keep only current month/year
    month_df = group[
        (group["Date_temp"].dt.month == curr_month) &
        (group["Date_temp"].dt.year == curr_year)
    ].copy()

    if month_df.empty:
        return pd.Series(["Instock"] * len(group), index=group.index)

    latest_row = month_df.iloc[-1]
    latest_date = latest_row["Date_temp"]
    latest_inv = latest_row["Available Inventory"]

    # ---------- CASE 1 : CURRENT INVENTORY = 0 ----------
    if latest_inv == 0:

        prev_positive = month_df[
            (month_df["Date_temp"] < latest_date) &
            (month_df["Available Inventory"] > 0)
        ]

        if not prev_positive.empty:
            oos_date = prev_positive.iloc[-1]["Date_temp"] + pd.Timedelta(days=1)
        else:
            zero_rows = month_df[month_df["Available Inventory"] == 0]
            oos_date = zero_rows.iloc[0]["Date_temp"]

    # ---------- CASE 2 : CURRENT INVENTORY > 0 ----------
    else:
        sell_through = latest_row["sell_through"]

        if pd.notna(sell_through) and sell_through > 0:
            days = int(latest_inv // sell_through)
            oos_date = latest_date + pd.Timedelta(days=days)
        else:
            oos_date = "Instock"

    return pd.Series([oos_date] * len(group), index=group.index)


# Apply per ASIN
inv_df["OOS Date"] = inv_df.groupby("ASIN", group_keys=False).apply(calculate_oos)

# Convert OOS Date to same format (if it's a date)
inv_df["OOS Date"] = inv_df["OOS Date"].apply(
    lambda x: x.strftime("%d-%b-%Y") if isinstance(x, pd.Timestamp) else x
)

# Remove temporary column
inv_df.drop(columns=["Date_temp"], inplace=True)
```

## Save correct format of revenue page

```python
import pandas as pd
from datetime import datetime

file_path = r"C:\Users\Shubham Verma\OneDrive - Incircle Ecoproducts Pvt Ltd\Thrivebrands - Client Database\Pattex\vendor_central\uae\sales\output\pattex_revenue.csv"

# Read existing CSV
df = pd.read_csv(file_path)

previous_rows = len(df)

# Create temp date column (DO NOT TOUCH ORIGINAL)
df['_temp_date'] = pd.to_datetime(df['Date'], format='%d-%b-%Y', errors='coerce').dt.normalize()

# Today's normalized date
today = pd.Timestamp.today().normalize()
current_day = today.day

# First day of current month (no time)
first_day_current_month = today.replace(day=1)

# Apply logic
if current_day <= 3:
    # Keep only <= last day of previous-to-previous month
    cutoff_date = first_day_current_month - pd.DateOffset(months=1)
    cutoff_date = cutoff_date.replace(day=1) - pd.DateOffset(days=1)

    df = df[df['_temp_date'] <= cutoff_date]

else:
    # Remove current month data
    df = df[df['_temp_date'] < first_day_current_month]

# Metrics after deletion
current_rows = len(df)
rows_deleted = previous_rows - current_rows

# Prepare merged_df
merged_df['_temp_date'] = pd.to_datetime(merged_df['Date'], format='%d-%b-%Y', errors='coerce').dt.normalize()

rows_appended = len(merged_df)

# Append new data
df = pd.concat([df, merged_df], ignore_index=True)

final_rows = len(df)

# Drop temp column before saving
df.drop(columns=['_temp_date'], inplace=True)

# Save back
df.to_csv(file_path, index=False)

# Print metrics
print(f"Previous rows: {previous_rows}")
print(f"Current rows: {current_rows}")
print(f"Rows deleted: {rows_deleted}")
print(f"Rows appended: {rows_appended}")
print(f"Final rows: {final_rows}")
```

```python
import os
import time
import random
import datetime
import pandas as pd

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
import selenium.common.exceptions as se

# ==============================
# INPUT FILE
# ==============================

INPUT_FILE = r"C:\Users\Shubham Verma\OneDrive - Incircle Ecoproducts Pvt Ltd\Thrivebrands - Client Database\Pattex\vendor_central\uae\BuyBox\input\pattex_buybox_asins.xlsx"

DOMAIN = "ae"
CURRENCY = "AED"

# ==============================
# DATE
# ==============================

now = datetime.datetime.now()
run_date = now.strftime("%Y-%m-%d")
run_time = now.strftime("%H:%M:%S")

# ==============================
# LOAD ASINS
# ==============================

df = pd.read_excel(INPUT_FILE)
asins = df["ASIN"].dropna().astype(str).unique()
print("Total ASINs:", len(asins))

# ==============================
# SELENIUM SETUP
# ==============================

options = Options()
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--start-maximized")
options.add_argument("--disable-gpu")
options.add_argument("--no-sandbox")
options.add_experimental_option("excludeSwitches", ["enable-automation"])
options.add_experimental_option('useAutomationExtension', False)

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# Hide webdriver detection
driver.execute_script("Object.defineProperty(navigator, 'webdriver', {get: () => undefined})")

results = []

# ==============================
# PRICE FUNCTION (PDP)
# ==============================

def get_price():
    try:
        whole = driver.find_element(By.CLASS_NAME, "a-price-whole").text
        frac = driver.find_element(By.CLASS_NAME, "a-price-fraction").text
        return f"{whole}.{frac}".replace(",", "")
    except:
        return ""

# ==============================
# MOQ FUNCTION
# ==============================

def get_moq():
    try:
        text = driver.find_element(
            By.XPATH,
            "//*[contains(text(),'Minimum order quantity')]"
        ).text
        moq = text.split(":")[-1].strip()
        return moq
    except:
        return "1"

# ==============================
# SCRAPE LOOP
# ==============================

for asin in asins:
    print("\n" + "="*50)
    print(f"Processing: {asin}")
    print("="*50)

    # ==============================
    # PRODUCT PAGE (PDP)
    # ==============================

    url = f"https://www.amazon.{DOMAIN}/dp/{asin}"

    try:
        driver.get(url)
    except se.TimeoutException as e:
        print(f"[ERROR] Timeout loading PDP for {asin}, skipping ASIN. {e}")
        continue

    time.sleep(random.uniform(4, 6))

    # PRODUCT TITLE
    try:
        title = WebDriverWait(driver, 20).until(
            EC.presence_of_element_located((By.ID, "productTitle"))
        ).text.strip()
        print(f"Title: {title[:60]}...")
    except:
        print(f"[WARN] Could not find productTitle for {asin}, skipping ASIN.")
        continue

    # PRICE (PDP)
    price = get_price()

    # ==============================
    # SELLER (BUYBOX) - ROBUST
    # ==============================

    seller = "Amazon.ae"

    # 1) Classic seller profile trigger
    try:
        seller_el = driver.find_element(By.ID, "sellerProfileTriggerId")
        txt = seller_el.text.strip()
        if txt:
            seller = txt
    except:
        pass

    # 2) Generic "Sold by" pattern (anchor)
    if not seller or seller == "Amazon.ae":
        try:
            seller_el = driver.find_element(
                By.XPATH,
                "//*[contains(normalize-space(text()),'Sold by')]/following::a[1]"
            )
            txt = seller_el.text.strip()
            if txt:
                seller = txt
        except:
            pass

    # 3) Tabular buybox pattern
    if not seller or seller == "Amazon.ae":
        try:
            seller_el = driver.find_element(
                By.XPATH,
                "//div[@id='tabular-buybox']//span[contains(@class,'tabular-buybox-text') and contains(.,'Sold by')]/following::a[1]"
            )
            txt = seller_el.text.strip()
            if txt:
                seller = txt
        except:
            pass

    # 4) Merchant info pattern
    if not seller or seller == "Amazon.ae":
        try:
            seller_el = driver.find_element(
                By.XPATH,
                "//div[@id='merchant-info']//a[contains(@href,'seller')]"
            )
            txt = seller_el.text.strip()
            if txt:
                seller = txt
        except:
            pass

    seller = seller.strip() if seller else "Amazon.ae"
    print(f"BuyBox Seller: {seller}")

    # MOQ
    moq = get_moq()

    # SAVE BUYBOX ROW
    results.append({
        "Date": run_date,
        "Time": run_time,
        "ASIN": asin,
        "Title": title,
        "Seller": seller,
        "Price": price,
        "MOQ": moq,
        "BuyBox": "YES",
        "Link": url
    })

    buybox_seller = seller

    # ==============================
    # OFFER LISTING PAGE (AOD) WITH RETRIES
    # ==============================

    offer_url = f"https://www.amazon.{DOMAIN}/gp/offer-listing/{asin}"
    max_retries = 3
    offers_loaded = False

    for attempt in range(1, max_retries + 1):
        try:
            print(f"\n[INFO] Loading offer page (attempt {attempt}/{max_retries})...")
            driver.get(offer_url)
            
            # Wait for offer container to load
            try:
                WebDriverWait(driver, 15).until(
                    EC.presence_of_element_located((By.ID, "aod-container"))
                )
                print("[INFO] Offer container loaded")
            except:
                print("[WARN] Offer container not found, trying alternate selectors")
            
            time.sleep(random.uniform(3, 5))
            offers_loaded = True
            break
            
        except (se.TimeoutException, se.WebDriverException) as e:
            print(f"[WARN] Timeout loading offers for {asin} (attempt {attempt}/{max_retries}): {e}")
            if attempt < max_retries:
                time.sleep(5)
            else:
                print(f"[ERROR] Skipping offers for {asin} after {max_retries} failed attempts.")
                offers_loaded = False

    if not offers_loaded:
        continue

    # ==============================
    # FIND OFFERS - MULTIPLE APPROACHES
    # ==============================

    offers = []
    
    # Approach 1: Standard aod-offer divs
    offers_approach1 = driver.find_elements(By.XPATH, "//div[contains(@id,'aod-offer')]")
    if offers_approach1:
        print(f"[INFO] Found {len(offers_approach1)} offers using approach 1")
        offers = offers_approach1
    
    # Approach 2: aod-offer-list divs
    if not offers:
        offers_approach2 = driver.find_elements(By.XPATH, "//div[@id='aod-offer-list']//div[contains(@class,'aod-offer')]")
        if offers_approach2:
            print(f"[INFO] Found {len(offers_approach2)} offers using approach 2")
            offers = offers_approach2
    
    # Approach 3: Generic offer containers
    if not offers:
        offers_approach3 = driver.find_elements(By.XPATH, "//div[contains(@class,'a-section aod-offer')]")
        if offers_approach3:
            print(f"[INFO] Found {len(offers_approach3)} offers using approach 3")
            offers = offers_approach3

    # Approach 4: pinned-offer and other offers
    if not offers:
        offers_approach4 = driver.find_elements(By.XPATH, "//div[@id='aod-pinned-offer'] | //div[contains(@id,'aod-offer-')]")
        if offers_approach4:
            print(f"[INFO] Found {len(offers_approach4)} offers using approach 4")
            offers = offers_approach4

    if not offers:
        print(f"[WARN] No offers found for {asin} on offer listing page")
        
        # Debug: Save page source
        try:
            debug_file = f"debug_{asin}.html"
            with open(debug_file, "w", encoding="utf-8") as f:
                f.write(driver.page_source)
            print(f"[DEBUG] Page source saved to {debug_file}")
        except:
            pass
        continue

    print(f"[INFO] Processing {len(offers)} offers...")
    
    seen_sellers = set()
    offers_count = 0

    for idx, offer in enumerate(offers):
        try:
            seller = ""

            # 1) Seller as link - anchor with href to seller profile or seller page
            links = offer.find_elements(
                By.XPATH,
                ".//div[contains(@id,'aod-offer-soldBy') or contains(@class,'soldBy')]//a[not(contains(@href,'/product-reviews/'))]"
            )
            if links:
                txt = links[0].text.strip()
                if txt and txt != "":
                    seller = txt

            # 2) If link text empty, try title attribute
            if (not seller) and links:
                title_attr = (links[0].get_attribute("title") or "").strip()
                if title_attr:
                    seller = title_attr

            # 3) Fallback: soldBy container text
            if not seller:
                try:
                    sold_by_containers = offer.find_elements(
                        By.XPATH,
                        ".//div[contains(@id,'aod-offer-soldBy') or contains(@class,'soldBy')]"
                    )
                    if sold_by_containers:
                        txt = (sold_by_containers[0].get_attribute("innerText") or "").strip()
                        if txt:
                            lines = [l.strip() for l in txt.split("\n") if l.strip()]
                            # Filter out rating/feedback lines
                            clean_lines = [
                                l for l in lines
                                if "positive over last 12 months" not in l.lower()
                                and "positive in the last 12 months" not in l.lower()
                                and "feedback" not in l.lower()
                                and "ratings" not in l.lower()
                                and "sold by" not in l.lower()
                                and "%" not in l
                            ]
                            if clean_lines:
                                seller = clean_lines[0]
                except:
                    pass

            # 4) Try merchant-info if available
            if not seller:
                try:
                    merchant = offer.find_element(
                        By.XPATH,
                        ".//div[contains(@id,'merchant-info')]//a"
                    )
                    seller = merchant.text.strip()
                except:
                    pass

            seller = seller.strip()

            if not seller:
                print(f"  [WARN] Offer {idx+1}: Could not extract seller name")
                continue

            # PRICE on offer row
            price = ""
            try:
                # Try whole and fraction separately
                whole = offer.find_element(By.CSS_SELECTOR, ".a-price-whole").text
                frac = offer.find_element(By.CSS_SELECTOR, ".a-price-fraction").text
                price = f"{whole}.{frac}".replace(",", "")
            except:
                try:
                    # Try getting full price text
                    price_elem = offer.find_element(By.CSS_SELECTOR, ".a-price .a-offscreen")
                    price = price_elem.get_attribute("textContent").strip().replace(CURRENCY, "").strip()
                except:
                    try:
                        # Alternative price selector
                        price_elem = offer.find_element(By.XPATH, ".//span[contains(@class,'a-price')]//span[@class='a-offscreen']")
                        price = price_elem.get_attribute("textContent").strip().replace(CURRENCY, "").strip()
                    except:
                        pass

            # Skip empty price
            if not price:
                print(f"  [WARN] Offer {idx+1}: {seller} - No price found")
                continue

            # Skip BuyBox seller
            if seller.lower() == buybox_seller.lower():
                print(f"  [SKIP] Offer {idx+1}: {seller} (BuyBox seller)")
                continue

            # Skip duplicates
            if seller.lower() in [s.lower() for s in seen_sellers]:
                print(f"  [SKIP] Offer {idx+1}: {seller} (duplicate)")
                continue

            seen_sellers.add(seller)
            offers_count += 1

            results.append({
                "Date": run_date,
                "Time": run_time,
                "ASIN": asin,
                "Title": title,
                "Seller": seller,
                "Price": price,
                "MOQ": "1",
                "BuyBox": "NO",
                "Link": url
            })

            print(f"  [✓] Offer {idx+1}: {seller} - {CURRENCY} {price}")

        except Exception as e:
            print(f"  [ERROR] Offer {idx+1}: {e}")
            pass

    print(f"\n[SUMMARY] {asin}: BuyBox + {offers_count} other sellers = {offers_count + 1} total")

# ==============================
# CLOSE BROWSER
# ==============================

driver.quit()

# ==============================
# SAVE OUTPUT
# ==============================

df_out = pd.DataFrame(results)
# output_file = r"C:\Users\Shubham Verma\OneDrive - Incircle Ecoproducts Pvt Ltd\Thrivebrands - Client Database\Pattex\vendor_central\uae\BuyBox\output\amazon_buybox_scrape_result.xlsx"
# df_out.to_excel(output_file, index=False)
# print(f"\n{'='*50}")
# print(f"SCRAPING COMPLETE!")
# print(f"{'='*50}")
# print(f"Total rows collected: {len(results)}")
# print(f"Saved to: {output_file}")
```
